```python
#5. train the YOLO model 
from ultralytics import YOLO

# Load a pretrained YOLOv8 model
model = YOLO("yolov8n.pt")

# Train on your dataset
results = model.train(
    data="data.yaml",   # path to YAML file
    epochs=2,          # number of epochs
    imgsz=124,          # image size
    batch=4,
    name="defect_detector",
    project="runs/train"
)

print("✅ Training Complete! Best weights saved in 'runs/train/defect_detector/weights/best.pt'")
```
