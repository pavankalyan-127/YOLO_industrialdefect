```python
# pretrained model using COCO dataset (80 object classes: car, person, bottle, dog,)
# ehatever the object present inside the coco pre trained it will give results other whise it will gove only output no class to it 

from ultralytics import YOLO
import cv2

# Load pretrained YOLOv8 model (nano version for speed)
model = YOLO('yolov8n.pt')

# Load image
image_path = r"C:\Users\pavan\Downloads\DETR_Obg\data\test_images\test2.img.jpeg"
results = model(image_path, show=True)  # show=True will open OpenCV window

# Save the result
for r in results:
    annotated = r.plot()
    cv2.imwrite("output_detected.jpg", annotated)
print("✅ Detection completed! Saved as output_detected.jpg")
```

```python
# using on video to detect objects

from ultralytics import YOLO
import cv2

# Load model
model = YOLO("yolov8s.pt")  # use 'yolov8n.pt' for speed, 'yolov8x.pt' for accuracy

# Open video
cap = cv2.VideoCapture("input_video.mp4")

if not cap.isOpened():
    print("❌ Error: Cannot open video file")
    exit()

# Output video settings
out = cv2.VideoWriter(
    "output_video.mp4",
    cv2.VideoWriter_fourcc(*'mp4v'),
    cap.get(cv2.CAP_PROP_FPS),
    (int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)), int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)))
)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Run detection on the frame
    results = model(frame)
    annotated_frame = results[0].plot()

    out.write(annotated_frame)
    cv2.imshow("YOLOv8 Detection", annotated_frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
out.release()
cv2.destroyAllWindows()
print("✅ Detection finished. Output saved as output_video.mp4")
```

```python
# live camers detection

from ultralytics import YOLO
import cv2

# Load pretrained YOLOv8 model
model = YOLO("yolov8n.pt")

cap = cv2.VideoCapture(0)  # 0 = webcam

while True:
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame)
    annotated = results[0].plot()

    cv2.imshow("YOLOv8 Live Detection", annotated)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
